```python
#splitting the dataset
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
X=sm.add_constant(ipl_data_encoded)
Y=ipl_data['SOLD PRICE']
train_X,test_X,train_Y,test_Y=train_test_split(X,Y,train_size=0.8,random_state=42)
```

```python
#building the model on the tarining dataset
ipl_model=sm.OLS(train_Y,train_X).fit()
ipl_model.summary()
```

```python
#checking for multi-collinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor
def get_vif_factors(X):
    X_matrix=X.values
    vif=[variance_inflation_factor(X_matrix,i) for i in range (X_matrix.shape[1])]
    vif_factors=pd.DataFrame()
    vif_factors['column']=X.columns
    vif_factors['VIF']=vif
    return vif_factors
```

```python
vif_factors=get_vif_factors(X[X_features])
vif_factors
```

```python
#checking correlation of columns with large VIFs
columns_with_large_vif=vif_factors[vif_factors.VIF>4].column
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(20,18))
sn.heatmap(X[columns_with_large_vif].corr(), annot=True);
plt.title("Heatmap depicting correlation between features");
plt.show()
```

```python
columns_to_be_removed=["T-RUNS","T-WKTS","RUNS-S","HIGH SCORE","AVE","RUNS-C","SR-B","AVE-BL","ECON","ODI-SR-B","ODI-RUNS-S","AGE_2","SR-BL"]
```

```python
X_new_features=list(set(X_features)-set(columns_to_be_removed))
```

```python
get_vif_factors(X[X_new_features])
```

```python
train_X=train_X[X_new_features]
ipl_model_2=sm.OLS(train_Y,train_X).fit()
ipl_model_2.summary()
```

```python
significant_var=['COUNTRY_IND','COUNTRY_ENG','SIXERS','CAPTAINCY EXP_1']
train_X=train_X[significant_var]
ipl_model_3=sm.OLS(train_Y,train_X).fit()
ipl_model_3.summary()
```

```python
#residual analysis
#test for normality of residuals(P-P Plot)
ipl_model_3_residual=ipl_model_3.resid
probplot=sm.ProbPlot(ipl_model_3_residual)
plt.figure(figsize=(8,6))
probplot.ppplot(line='45')
plt.title("Normal P-P Plot")
plt.show()
```

```python
#residual plot for HOMOSCEDASTICITY
def get_std_val(vals):
    return (vals-vals.mean())/vals.std()
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
plt.scatter(get_std_val(ipl_model_3.fittedvalues),get_std_val(ipl_model_3_residual))
plt.title("residual plot");
plt.xlabel("std pred val");
plt.ylabel("std residual");
```
